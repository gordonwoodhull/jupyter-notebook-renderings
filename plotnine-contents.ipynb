```python
#| echo: false
from quarto import theme_brand_plotnine

light_theme = theme_brand_plotnine('light-brand.yml')
dark_theme = theme_brand_plotnine('dark-brand.yml')
```

```python
from plotnine import *
import plotly.express as px

gapminder = px.data.gapminder()


def gapminder_plot(year):
    """
    Create a gapminder scatter plot using plotnine

    Parameters:
    year: int - Year to filter data for
    theme: str - plotnine theme name (options: 'default', 'minimal', 'classic',
                 'gray', 'dark', 'void', 'light', 'linedraw', 'bw')
    """
    # Filter data for the specified year (using pandas DataFrame)
    gapminder_year = gapminder[gapminder['year'] == year]

    return (
        ggplot(gapminder_year, aes(x="gdpPercap", y="lifeExp"))
        + geom_point(aes(color="continent", size="pop"), alpha=0.7)
        + scale_x_log10(name="GDP per Capita")
        + scale_y_continuous(name="Life Expectancy")
        + scale_color_discrete(name="Continent")
        + scale_size_continuous(
            name="Population",
            range=(2, 20),  # Equivalent to size range in Altair
            guide=guide_legend(override_aes={"alpha": 1, "color": "steelblue"})
        )
        + labs(title=f"Gapminder Data for {year}")
        + theme_minimal()
    )
```

```python
#| label: gapminder-1957
#| renderings: [light, dark]
plot1957 = gapminder_plot(1957)
display(plot1957 + light_theme)
display(plot1957 + dark_theme)
```
