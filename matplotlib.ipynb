```python
#| label: matplotlib-normal
#| renderings: [light,dark]
#| echo: false
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 0
std_dev = 1

# Generate data
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Plotting
light_style()
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Normal Distribution')
plt.title('Normal Distribution Curve')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

dark_style()
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Normal Distribution')
plt.title('Normal Distribution Curve')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#| label: matplotlib-line
#| renderings: [light,dark]
#| echo: false
import matplotlib.pyplot as plt

light_style()
plt.title("Hello")
plt.plot([1,2,3])
plt.grid(True)
plt.show(block=False)

dark_style()
plt.figure()
plt.title("Hello")
plt.plot([1,2,3])
plt.grid(True)
plt.show(block=False)
```

```python
#| label: matplotlib-sine
#| renderings: [light,dark]
#| echo: false

# author: "anthropic claude-3-5-sonnet-20240620"
import numpy as np
import matplotlib.pyplot as plt

# Generate data points
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

light_style()
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.show()

dark_style()
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.show()
```
